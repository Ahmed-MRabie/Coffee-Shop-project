```python
class CoffeeShop:
    
    def __init__(self, coffee_name, menu_dict, orders_list):
        """Initialize attributes to describe the Coffee Shop"""
        self.coffee_name = coffee_name
        self.menu_dict = menu_dict
        self.orders_list = orders_list

    #=======================================================================================================
    def add_order(self, item_name):
        """Adds the name of the item to the end of the orders list if it exists on the menu"""
        for item in self.menu_dict:
            
            if item_name.lower() in item["name"].lower():
                self.orders_list.append(item_name.title())
                return "Order added!"
                
        return "Sorry!, This item is currently unavailable!"

    #=======================================================================================================
    def list_orders(self):
        """Returns the item names of the orders taken"""
        return f"all items of the current order:\n {self.orders_list}"

    #=======================================================================================================
    def fulfill_order(self):
        """Return the ready item in the list orders"""
        if len(self.orders_list) == 0:
            return "All orders have been fulfilled!"
        else:
            return (f"The {self.orders_list.pop(0)} is ready!")

    #=======================================================================================================
    def due_amount(self):
        """Returns the total amount due for the orders taken"""
        total_amount = 0
        
        for order_item in self.orders_list:
            for menu_item in self.menu_dict:
                if order_item == menu_item["name"]:
                    total_amount += menu_item["price"]

        return total_amount

    #=======================================================================================================
    def cheapest_item(self):
        """Returns the name of the cheapest item on the menu"""
        cheap_item = self.menu_dict[0]
        cheap_item_name = cheap_item["name"]
        
        for menu_item in self.menu_dict:
            if cheap_item["price"] > menu_item["price"]:
                cheap_item = menu_item
                cheap_item_name = menu_item["name"]

        return cheap_item_name
    #------------------------------------------------------------------
    def cheapest_item_list(self):
        """Returns the name of the cheapest item on the menu"""
        cheap_items_name = []
        cheap_items_price = []
        cheap_item = self.menu_dict[0]
        cheap_items_name.append(cheap_item["name"])
        cheap_items_price.append(cheap_item["price"])
        
        for menu_item in self.menu_dict[1:]:
            if cheap_items_price[0] == menu_item["price"]:
                cheap_items_name.append(menu_item["name"])
                cheap_items_price.append(menu_item["price"])
                
            if cheap_items_price[0] > menu_item["price"]:
                cheap_items_name.clear()
                cheap_items_price.clear()
                cheap_items_name.append(menu_item["name"])
                cheap_items_price.append(menu_item["price"])
                
        return cheap_items_name

    #=======================================================================================================
    def drinks_only(self):
        """Returns only the item names of type drink from the menu"""
        drinks_items = []
        
        for item in self.menu_dict:
            if item["type"].lower() == "drink":
                drinks_items.append(item["name"])

        return drinks_items

    #=======================================================================================================
    def food_only(self):
        """Returns only the item names of type drink from the menu"""
        food_items = []
        
        for item in self.menu_dict:
            if item["type"].lower() == "food":
                food_items.append(item["name"])

        return food_items
```
